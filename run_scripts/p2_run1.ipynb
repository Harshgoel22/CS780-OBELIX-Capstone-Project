```python
import torch
checkpoint = torch.load("../model_phase2_sub1/ppo_checkpoint.pt", map_location="cpu")
ppo_results = checkpoint["results"]
```

```python
steps_list = []
rewards_list = []

for (gs, er) in ppo_results:
    steps_list.append(gs)
    rewards_list.append(er)
```
